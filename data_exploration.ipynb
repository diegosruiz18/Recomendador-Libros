# Exploración de datos

En este notebook realizaremos el preprocesamiento del dataset, análisis y tratamiento de valores nulos, creación de nuevas columnas, eliminar columnas inncesarias, etc, con el fin de dejar los datos listos para los siguientes pasos.

```python
#Descargando data de Kaggle
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
#Importando librerías
import pandas as pd
```

```python
libros = pd.read_csv(f"{path}/books.csv") #dataset
```

```python
libros
```

La columna de mayor interés es "description" dado que el recomendador de libros trabajará con esta

```python
#Librerías para visualización de gráficos
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Valores nulos
ax = plt.axes()
sns.heatmap(libros.isna().transpose(), cbar=False, ax=ax)
#Labels
plt.xlabel("Filas")
plt.ylabel("Columnas")
plt.show()
```

La columna "subtitle" presenta muchos valores nulos, se repite un patrón en los valores nulos de las columnas "average_rating", "num_pages" y "ratings_count".

```python
#Creando nuevas columnas
import numpy as np

libros["missing_description"] = np.where(libros["description"].isna(),1,0) #si descripcion es nulo colocar 1, de lo contrario 0
libros["age_of_book"] = 2025 - libros["published_year"] #obteniendo edad del libro
```

```python
#Separando columnas relevantes en una lista
columnas_de_interes = ["num_pages", "age_of_book", "missing_description", "average_rating"]

#Matriz de correlación
correlation_matrix = libros[columnas_de_interes].corr(method="spearman")

#Mapa de calor
sns.set_theme(style="white")
plt.figure(figsize = (8,6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm",
                      cbar_kws={"label": "Correlación Spearman"})
heatmap.set_title("Mapa de calor")
plt.show()
```

La correlaciones entre "missing_description" y las demás columnas son débiles lo que indica que no hay relación aparente.
Esto significa que las descripciones son bastante representativas respecto a estas dimensiones, por ejemplo: libros grandes o pequeños, nuevos o antiguos, etc, en el mayor de los casos cuentan con su respectiva descripción. Esto ofrece una mayor variedad para el recomendador de libros.

```python
libros[(libros["description"].isna()) |
       (libros["num_pages"].isna()) |
       (libros["average_rating"].isna()) |
       (libros["published_year"].isna())] #shape
```

Las filas con valores nulos son 303, menos del 5%, por lo que se procede a eliminarlas.

```python
#Eliminando valores nulos
libros_no_faltantes = libros[~(libros["description"].isna()) &
                             ~(libros["num_pages"].isna()) &
                             ~(libros["average_rating"].isna()) &
                             ~(libros["published_year"].isna())]
```

```python
libros_no_faltantes
```

Ahora procedemos a tratar la columna "categories" ya que tiene muchos valores distintos (531) lo cual no es óptimo para el recomendador.

```python
#Revisando distribución de categorías
libros_no_faltantes["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

Observamos que muchas categorías solo tienen 1 o 2 recuentos.

Examinamos las descripciones, estas deben ser significativas, es decir proveer suficiente información acerca del libro para que sean útiles.

```python
#Creando nueva columna que indica el numero de palabras de la descripcion
libros_no_faltantes["words_in_description"] = libros_no_faltantes["description"].str.split().str.len()
libros_no_faltantes
```

Se visualiza una distribución por lo general uniforme de los datos, entre un rango de 20 y 320 palabras en descripción.

```python
#Filtrando por cantidad de palabras en descripcion
#Entre 1 y 4 palabras
libros_no_faltantes.loc[libros_no_faltantes["words_in_description"].between(1, 4), "description"]
```

```python
#Entre 5 y 14 palabras
libros_no_faltantes.loc[libros_no_faltantes["words_in_description"].between(5, 14), "description"]
```

```python
#Entre 15 y 24 palabras
libros_no_faltantes.loc[libros_no_faltantes["words_in_description"].between(15, 24), "description"]
```

```python
#Entre 25 y 34 palabras
libros_no_faltantes.loc[libros_no_faltantes["words_in_description"].between(25, 34), "description"]
```

A partir de 25 palabras se puede considerar una descripción util.

```python
#Filtrando filas que tengan al menos 25 palabras en descripción
libros_no_faltantes_25_palabras = libros_no_faltantes[libros_no_faltantes["words_in_description"] >= 25]
libros_no_faltantes_25_palabras
```

```python
#Creando una nueva columna fusionando el titulo y subtitulo si este ultimo existe
libros_no_faltantes_25_palabras.loc[:, "title_and_subtitle"] = np.where(
    libros_no_faltantes_25_palabras["subtitle"].isna(),
    libros_no_faltantes_25_palabras["title"],
    libros_no_faltantes_25_palabras[["title", "subtitle"]].astype(str).agg(": ".join, axis=1)
)

libros_no_faltantes_25_palabras
```

```python
#Creando columna que combina "isbn13" y "description"
libros_no_faltantes_25_palabras.loc[:,"tagged_description"] = libros_no_faltantes_25_palabras[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)

libros_no_faltantes_25_palabras
```

```python
#Elimnando columnas innecesarias y exportando archivo .csv
(libros_no_faltantes_25_palabras.drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1)
                                .to_csv("books_cleaned.csv", index=False)
)
```
